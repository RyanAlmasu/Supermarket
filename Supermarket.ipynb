# Supermarket_Datasets

## Importing Data using pandas

```python
import pandas as pd
import matplotlib.pyplot as plt
data_supermarket = pd.read_csv("supermarket.csv")
data_supermarket.head()
```

#### Showing Total Gross Income from Data

```python
total_gross_income = data_supermarket["gross income"].sum()
dollar = str(" $")
desc = str("This is the total gross income in the data is ")
print(f'{desc}{total_gross_income}{dollar}')
```

#### Showing best all values from data

```python
rate_product = data_supermarket["Rating"].max() and data_supermarket.max()
rate_product
```

#### Showing worst values in data

```python
min_rating = data_supermarket["Rating"].min() and data_supermarket.min()
min_rating
```

## Showing best adverstiment on product based in hour we get on data

#### Converting date and time

```python
df = data_supermarket
df["DateTime"] = df['Date'] + " " + df['Time'] 
df["DateTime"] = pd.to_datetime(df["DateTime"])

df["Hour"] = df["DateTime"].dt.hour
df["Minute"] = df["DateTime"].dt.minute
df["Count"] = 1
df

```

#### Visualizing Data on the best hour to put ads

```python
hours = [hour for hour , df in data_supermarket.groupby("Hour")]
quantity_salesinhour = df.groupby("Hour").sum()
plt.bar(hours , quantity_salesinhour["Quantity"] , color='Orange')
plt.xticks(hours)
plt.ylabel("Product Sales")
plt.xlabel("Hours")
plt.grid()
plt.show()
```

#### So the best adverstisment we can do  is around 19 pm

## Most many Products Sold Based in Quantity

#### The most product sold 

```python
product_group = data_supermarket.groupby("Product line")
total_quantity =  product_group.sum()["Quantity"]
print(total_quantity)
```

#### Visualizing the Data most product sold

```python
products = [product for product , df in product_group]
plt.plot(products , total_quantity , color ='Red')
plt.xticks(products , rotation="vertical", size=10)
plt.xlabel("Product")
plt.ylabel("Quantity")
plt.grid()
plt.show()
```

## The best month for sale 

#### Converting datetime into month 

```python
df["Month"] = df["DateTime"].dt.month
df.head()
```

#### Find the most value of month for product sales in gross income

```python
best_value_month = df.groupby("Month").sum()
month = df["Month"].unique()
plt.bar(month, best_value_month["gross income"] , color='Green')
plt.xticks(range(1,4))
plt.grid()
plt.xlabel("Months")
plt.ylabel("Gross Income")
plt.show()
```

## Find the most income product sales (Total income + Taxes)

```python
product_month_sales = df.groupby("Product line").sum()
product_line = df["Product line"].unique()
plt.bar(product_line, product_month_sales["Total"])
plt.xticks(rotation='vertical')
plt.xlabel("Products")
plt.ylabel('Total Income $')
plt.grid()
plt.show()
```

## Showing the most high value of the cities 

#### Visualizing data of the most high gross income each cities

```python
city = df["City"].unique()
sales_city = df.groupby("City").sum()
plt.plot(city, sales_city["gross income"], color="brown")
plt.grid()
plt.show()
```
